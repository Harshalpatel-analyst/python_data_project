How are in-demand skills trending for Data Analysts?

Methology

1.Aggregate skill counts monthly

2.Re-analyze based on percentage of total jobs

3.Plot the monthly skill demand

```python
import pandas as pd
import matplotlib.pyplot as plt
import ast
import seaborn as sns
```

```python
from datasets import load_dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda job_skills: ast.literal_eval(job_skills) if pd.notna(job_skills) else job_skills)
```

```python
df_DA_US = df[(df['job_title_short']=='Data Analyst')&(df['job_country']=='United States')].copy()
```

```python
df_DA_US['job_posted_month_no'] = df_DA_US['job_posted_date'].dt.month
```

```python
df_DA_US_explode = df_DA_US.explode('job_skills')
```

```python
df_DA_US_pivot = df_DA_US_explode.pivot_table(index= 'job_posted_month_no', columns= 'job_skills', aggfunc='size', fill_value = 0)
```

```python
df_DA_US_pivot.loc['total'] = df_DA_US_pivot.sum()
```

```python
df_DA_US_pivot = df_DA_US_pivot[df_DA_US_pivot.loc['total'].sort_values(ascending=False).index]
df_DA_US_pivot = df_DA_US_pivot.drop('total')
```

```python
df_DA_US_pivot
```

```python
df_DA_US_pivot.iloc[ : , :5].plot(kind = 'line')
plt.title('Trending Top Skills for Data Analyst in US')
plt.ylabel('Count')
plt.xlabel('')
plt.show()
```

```python
DA_totals = df_DA_US.groupby('job_posted_month_no').size()
```

```python
df_DA_US_percent = df_DA_US_pivot.iloc[:12].div(DA_totals/100, axis = 0)
df_DA_US_percent = df_DA_US_percent.reset_index()
df_DA_US_percent['job_posted_month'] = df_DA_US_percent['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_US_percent = df_DA_US_percent.set_index('job_posted_month')
df_DA_US_percent = df_DA_US_percent.drop(columns='job_posted_month_no')

df_DA_US_percent
```

```python
# divide first 12 rows of df_DA_pivot by DA_totals
df_DA_US_percent = df_DA_US_pivot.iloc[:12].div(DA_totals/100, axis=0)

# changes month number to month name
df_DA_US_percent = df_DA_US_percent.reset_index()
df_DA_US_percent['job_posted_month'] = df_DA_US_percent['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_US_percent = df_DA_US_percent.set_index('job_posted_month')
df_DA_US_percent = df_DA_US_percent.drop(columns='job_posted_month_no')
df_DA_US_percent
```

```python
from matplotlib.ticker import PercentFormatter

df_plot = df_DA_US_percent.iloc[:, :5]
sns.lineplot(data=df_plot, dashes=False, legend='full', palette='tab10')
sns.set_theme(style='ticks')
sns.despine() # remove top and right spines

plt.title('Trending Top Skills for Data Analysts in the US')
plt.ylabel('Likelihood in Job Posting')
plt.xlabel('2023')
plt.legend().remove()
plt.gca().yaxis.set_major_formatter(PercentFormatter(decimals=0))

# annotate the plot with the top 5 skills using plt.text()
for i in range(5):
    plt.text(11.2, df_plot.iloc[-1, i], df_plot.columns[i], color='black')

plt.show()
```

```python
from matplotlib.ticker import PercentFormatter, MultipleLocator

# Assuming df_DA_US_percent is already defined and contains the data
df_plot = df_DA_US_percent.iloc[:, :5]

# Plot the data
sns.lineplot(data=df_plot, dashes=False, legend='full', palette='tab10')
sns.set_theme(style='ticks')
sns.despine()  # remove top and right spines

# Set the title and labels
plt.title('Trending Top Skills for Data Analysts in the US')
plt.ylabel('Likelihood in Job Posting')
plt.xlabel('2023')

# Remove the legend
plt.legend().remove()

# Set the y-axis formatter to percentage and the major locator to 10% intervals
plt.gca().yaxis.set_major_formatter(PercentFormatter(decimals=0))
plt.gca().yaxis.set_major_locator(MultipleLocator(10))
plt.ylim(10, 60)

# Annotate the plot with the top 5 skills using plt.text()
# Apply an offset to avoid overlapping
offsets = [-1.5, 1.5, -1.5, 1.5, -1.5]  # Adjust these values as needed
for i in range(5):
    plt.text(11.2, df_plot.iloc[-1, i] + offsets[i], df_plot.columns[i], color='black')

# Show the plot
plt.show()
```
